```python
conn = sqlite3.connect('data/im.db')

cursor = conn.cursor()

# Retrieving all the table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")

# Fetching all results
tables = cursor.fetchall()

# Displaying the table names
print("Tables in the database:")
for table in tables:
    print(table[0])
```

```python
imdb = pd.read_sql(\
    """
    SELECT *
    FROM movie_basics
    JOIN movie_ratings 
        USING(movie_id) 
     
     """
    
    
    , conn)
imdb.head()
```

```python
imdb.isna().sum()
```

- runtime_minutes has 7620 of 73856 missing values. Less than 10% of the data. Drop missing rows

### THE MOVIEDB
COLS : genre_ids |	id |	original_language |	original_title |	popularity |	release_date |	title |	vote_average |	vote_count

```python
tmdb = pd.read_csv('data/tmdb.movies.csv', index_col=0)
tmdb.head()
```

```python
tmdb.info()
```

```python
tmdb.isna().sum()
```

- From the Movie db, drop genre_ids 

```python
tmdb = tmdb.drop(columns='genre_ids')
```

```python
# tmdb, imdb, tn - use tn later 
it_merged = pd.merge(imdb, tmdb, on='original_title', how='inner')
it_merged
```

```python
it_merged.isna().sum()
```

```python
# get average of ratings from imdb and tmdb 
# add the vote counts
it_merged['avg_rating'] = (it_merged['averagerating'] + it_merged['vote_average'])/2
it_merged['num_votes'] = it_merged['numvotes'] +  it_merged['vote_count']

# get release year from release date
it_merged['release_year'] = pd.to_datetime(it_merged['release_date']).dt.year
it_merged['release_date'] = pd.to_datetime(it_merged['release_date'])

it_merged
```

```python
# select the columns to keep.
cols = [
    'primary_title', 'original_title','title',
    'genres', 'original_language', 'runtime_minutes',
    'popularity', 'avg_rating', 'num_votes', 'release_date', 'release_year'
]
```

```python
it_merged = it_merged[cols]
it_merged
```

```python
it_merged.duplicated().sum()
```

```python
it_merged = it_merged.drop_duplicates()
```

```python
it_merged.duplicated().sum()
```

```python
tn.duplicated().sum()
```

```python
# rename tn_movie to original_title
tn = tn.rename(columns={'title':'original_title'})
tn
```

```python
tn.isna().sum()
```

```python
tn['production_budget'] = pd.to_numeric(tn['production_budget'].astype(str).str.replace('[\$,]', '', regex=True))
tn['domestic_gross'] = pd.to_numeric(tn['domestic_gross'].astype(str).str.replace('[\$,]', '', regex=True))
tn['worldwide_gross'] = pd.to_numeric(tn['worldwide_gross'].astype(str).str.replace('[\$,]', '', regex=True))

tn
```

```python
# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.hist(tn['production_budget'], bins=30, edgecolor='k')
plt.title('Production Budget Distribution')

plt.subplot(1, 3, 2)
plt.hist(tn['domestic_gross'], bins=30, edgecolor='k')
plt.title('Domestic Gross Distribution')

plt.subplot(1, 3, 3)
plt.hist(tn['worldwide_gross'], bins=30, edgecolor='k')
plt.title('Worldwide Gross Distribution')

plt.tight_layout()
plt.show()
```

```python
# Plot boxplots
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.boxplot(y=tn['production_budget'])
plt.title('Production Budget Boxplot')

plt.subplot(1, 3, 2)
sns.boxplot(y=tn['domestic_gross'])
plt.title('Domestic Gross Boxplot')

plt.subplot(1, 3, 3)
sns.boxplot(y=tn['worldwide_gross'])
plt.title('Worldwide Gross Boxplot')

plt.tight_layout()
plt.show()
```

### UNIVARIATE ANALYSIS

```python
# top 10 genres
it_merged['genres'].value_counts()[:10]
```

```python
fig, ax = plt.subplots(figsize=(10,8))

#x
x = it_merged['genres'].value_counts()[:10].index
#y
y = it_merged['genres'].value_counts()[:10].values


ax.barh(x,y)
ax.set_xlabel('count')
ax.set_ylabel('Genre')
ax.set_title('Distribution of top 10 Genres')
ax.invert_yaxis();
```

```python
# languages used in films
top5_lang = it_merged['original_language'].value_counts()[:5]
```

```python
# list of labels for the pie chart
labels = top5_lang.index

# # Create a pie chart
fig, ax = plt.subplots(figsize=(10, 8))
ax.pie(top5_lang, labels=labels, autopct='%1.1f%%')

ax.set_title('Top 5 Original Languages in Movies')

ax.legend(labels, title="Languages", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# # Display the pie chart
plt.show()



# ax.pie(top5_lang, autopct='%1.1f%%');
```
